# Take a look at the fractions

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.ascii as at
```

```python
from file_loc import FileLocs

fl = FileLocs()
file_base = fl.data_loc+'tmp/{type}_{prop}.npy'
```

```python
label1 = ['Cluster', 'Infall', 'Interlopers']
label2 = ['Cluster+Infall', 'Cluster+Interlopers', 'Infall+Interlopers']

def plot_fraction_pannel(method, prop, xscale=False, save=None):
    if method=='free_fall':
        xscale = True

    x = np.loadtxt(file_base.format(type=prop, prop=method)).T
    
    plt.clf()
    fig = plt.figure(figsize=(12,4))
    plt.subplot(1, 2, 1)
    
    count = np.zeros_like(x[0])
    for i in range(3):
        ii = 2*i + 1
        plt.errorbar(x[0], x[ii], yerr=x[ii+1], label=label1[i], fmt='o')
        
        #count = x[ii][0]
    print('count:',x[1])
    
    if xscale:
        plt.xscale('log')
    plt.legend(fontsize=12)
    #plt.xlabel(r'Log($M_\star/M_{\odot}$)',fontsize=16)
    plt.title('%s Fraction'%prop,fontsize=16)
    
    plt.subplot(1, 2, 2)
    for i in range(3):
        ii = (2*i + 1)+6
        plt.errorbar(x[0], x[ii], yerr=0.5*x[ii+1], label=label2[i], fmt='o')
    plt.legend(fontsize=12)
    #plt.xlabel(r'Log($M_\star/M_{\odot}$)',fontsize=18)
    plt.title('%s Fraction Excess'%prop,fontsize=16)
    if xscale:
        plt.xscale('log')
    del x
    
    if save is not None:
        plt.savefig(save, transparent=False)
        plt.clf()
```

```python
labels_mpr = ['sf', 'spiral', 's0', 'bulge', 'disk', 'bar', 'merger']
labels_bpt = ['lsf', 'liners', 'agn', 'compos', 'unclas']

labels = labels_mpr+labels_bpt
```

```python
dtype = 'radii'
for li in labels:
    outfile = '%s_%s.png'%(dtype,li)
    plot_fraction_pannel(dtype,li,save=outfile)

dtype = 'smass'
for li in labels:
    outfile = '%s_%s.png'%(dtype,li)
    plot_fraction_pannel(dtype,li,save=outfile)
```
